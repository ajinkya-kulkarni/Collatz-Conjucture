```python
import numpy as np
import matplotlib.pyplot as plt
import math

import random
```

```python
%config InlineBackend.figure_format ='retina'

plt.rcParams.update({'font.size': 20})
```

### Make an array of random integers

```python
a = int(1e1)

b = int(1e3)

input_array = np.arange(a, b)
```

### Define the function

```python
def collatz_sequence(value):
    
    sequence = []
    
    if value < 1:
        
        exit

    while value > 1:
        
        if value % 2 == 0:
            
            value = int(value / 2)
            
        else:
            
            value = int(3 * value + 1) 
            
        sequence.append(value)
    
    return np.asarray(sequence)
```

```python
collatz_sequence_array = []

for number in input_array:
    
    collatz_sequence_array.append([number, collatz_sequence(number)])
```

### Convergence time vs Input integer 

```python
convergence_time_array = []

for i in np.arange(len(input_array)):
    
    convergence_time_array.append([input_array[i], len(collatz_sequence_array[i][1])])
    
convergence_time_array = np.asarray(convergence_time_array)
```

### Plots

```python
plt.figure(figsize=(10, 6))

for i in np.arange(len(input_array)):
    
    plt.loglog(collatz_sequence_array[i][1], 
               linewidth = 5 * input_array[i]/max(input_array), 
               alpha = 0.5*input_array[i]/max(input_array))
    
plt.title('Convergence Time')

plt.xlabel(r'$t$')

plt.ylabel(r'$N$')

plt.xlim(1,)

plt.ylim(1,)

# plt.savefig('image1', dpi = 300, bbox_inches = 'tight')
```

```python
plt.figure(figsize=(10, 6))

plt.loglog(convergence_time_array[:, 0], convergence_time_array[:, 1], 'o', markersize = 10, alpha = 0.1)
    
plt.title('Convergence Time vs Integer')

plt.xlabel(r'$t$')

plt.ylabel(r'$N$')

plt.xlim(10,)

plt.ylim(a,)

# plt.savefig('image2', dpi = 300, bbox_inches = 'tight')
```
