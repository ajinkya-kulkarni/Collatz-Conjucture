```python
import numpy as np
import matplotlib.pyplot as plt
import math

import random
```

```python
%config InlineBackend.figure_format ='retina'

plt.rcParams.update({'font.size': 20})
```

### Make an array of random integers

```python
a = int(1e1)

b = int(1e3)

input_array = np.arange(a, b)
```

### Define the function

```python
def collatz_sequence(value):
    
    sequence = []
    
    if value < 1:
        
        exit

    while value > 1:
        
        if value % 2 == 0:
            
            value = int(value / 2)
            
        else:
            
            value = int(3 * value + 1) 
            
        sequence.append(value)
    
    return np.asarray(sequence)
```

```python
collatz_sequence_array = []

for number in input_array:
    
    collatz_sequence_array.append([number, len(collatz_sequence(number)), collatz_sequence(number)])
```

```python
collatz_sequence_times = []

for i in np.arange(len(input_array)):
    
    collatz_sequence_times.append(collatz_sequence_array[i][1])
```

### Plots

```python
plt.figure(figsize=(10, 6))

for i in np.arange(len(input_array)):
    
    plt.loglog(collatz_sequence_array[i][2], 
               linewidth = 5 * collatz_sequence_array[i][1]/max(collatz_sequence_times), 
               alpha = 0.5)

plt.xlabel(r'$t$')

plt.ylabel(r'$N$')

plt.xlim(1,)

plt.ylim(a,)

# plt.savefig('image1', dpi = 300, bbox_inches = 'tight')
```

```python
plt.figure(figsize=(10, 6))

for i in np.arange(len(input_array)):
    
    plt.scatter(collatz_sequence_array[i][0], collatz_sequence_array[i][1], 
                s = 200,
                alpha = 0.5 * collatz_sequence_array[i][1]/max(collatz_sequence_times))

plt.xlabel(r'$N$')

plt.ylabel(r'$t$')

plt.xlim(a, b)

plt.ylim(1,)

plt.yscale('log')

plt.xscale('log')

# plt.savefig('image1', dpi = 300, bbox_inches = 'tight')
```
